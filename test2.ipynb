```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
img = cv2.imread('notes.png')
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

```python
height, width = img.shape[:2]
print(height, width)
```

```python
white_height, black_height = [], []
```

```python
for i in range (width):
    flag = False
    white, black = 0, 0
    for j in range (height):
        value = img_gray[j, i]
        if value == 0:
            flag = True
            black += 1
            if white != 0:
                white_height.append(white)
                white = 0
        else:
            if flag:
                white += 1
            black_height.append(black)
            black = 0
```

```python
counter = Counter(white_height)
average = (counter.most_common(2)[0][0]*counter.most_common(2)[0][1] + counter.most_common(2)[1][0]*counter.most_common(2)[1][1])/(counter.most_common(2)[0][1]+counter.most_common(2)[1][1])
print(average*4)
```
